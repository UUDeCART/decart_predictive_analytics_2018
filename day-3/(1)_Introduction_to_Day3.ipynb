# Supervised Learning (Predictive learning)

#### Key Idea:

We have **[labeled data](https://en.wikipedia.org/wiki/Labeled_data)** about the subject. That is, we know some **truth** value and we want to learn how to predict this truth value.

## Approach

1. It is a task of inferring a function from labeled training data.
2. Each training record is an example or a [vector of features/attributes](https://en.wikipedia.org/wiki/Feature_vector) (predictors) and label (outcome) variables.
3. The inferred function maps the examples to new unseen example inferring the label.

# Model Development, Validation, and Test


## Steps
1. Remove duplicates
2. Shuffle data-- why?
2. Split data into training (60%), validation (a.k.a development) (20%) and test (20%).
3. Preprocessing data (usually the decision should be on training data only), such as standardization/normalization, regularization and/or feature selection/transformation
4. Model selection


## Bias-variance tradeoff               



***The two errors are the part of generalization error?***



1. Bias: the difference between prediction and actual value.
   - High Bias leads to model underfitting – too general to identify specific detailed data patterns.
2. Variance: the difference among predictions themselves.
   - High Variance leads to model overfitting – too specific to identify specific and sensitive to noise and outliers
